```python
# Library Management System
class Book:
  def __init__(self, title, author, genre):
    self.title = title
    self.author = author
    self.genre = genre
    self.is_available = True

  def display_info(self):
    print(f"Title: {self.title}")
    print(f"Author: {self.author}")
    print(f"Genre: {self.genre}")
    print("Status: Available" if self.is_available else "Status: Not Available")


class Library:

  def __init__(self):
    self.books = []

  # add new book to the library
  def add_book(self,book):
    self.books.append(book)

  def book_info(self):
    for book in self.books:
      book.display_info()

  # remove a book from the library - by title
  def remove_book(self,search_word):
    book_found = [book for book in self.books if search_word.lower() in book.title.lower()]
    print("Searching for the book.....")
    if book_found:
      self.books.remove(book_found)

  # search a book by title, author or genre
  def search_book(self,search_word):
    books_found = [book for book in self.books if search_word.lower() in book.title.lower() \
                   or search_word.lower() in book.author.lower() \
                   or search_word.lower() in book.genre.lower()]
    print("Searching for the book.....")
    if books_found:
      print("Search Result.")
      for book in books_found:
        book.display_info()
        print("-----------------------")
    else:
      print("No books found matching the search criteria.")

  # borrow a book
  def borrow_book(self,book_title):
    for book in self.books:
      if book.title.lower() == book_title.lower() and book.is_available:
        book.is_available = False
        return book
      return None

  # return a book
  def return_book(self,book,user):
    if book in user.borrowed_books:
      book.is_available = True
      user.borrowed_books.remove(book)
      print(f"{user.name} has returned '{book.title}'.")
    else:
      print("No book to return")

# ------------ User Class ------------------------
class User:

  def __init__(self,name):
    self.name = name
    self.borrowed_books = []

  def borrow_book(self, library, book_title):
    book = library.borrow_book(book_title)
    if book:
      self.borrowed_books.append(book)
      print(f"{self.name} has borrowed '{book.title}'.")
    else:
      print(f"'{book_title}' is not available in the library.")

  def return_book(self, library, book_title):
    for book in self.borrowed_books:
      if book.title.lower() == book_title.lower():
        library.return_book(book, self)
        return
    print("No book to return")


# ------------------
# create instance of the library class
library = Library()
users = []

while True:
  print("Welcome to the Library Management System")
  print("\nMenu:")
  print("1. Add a book")
  print("2. Create user")
  print("3. Search for a book")
  print("4. Borrow a book")
  print("5. Return a book")
  print("6. Exit")

  user_input = int(input("Please select an option from 1 to 6: "))

  if user_input == 1:
    # Adding new book to the library
    book_title = input("Please enter the title of the book: ").title()
    author = input("Please enter the name of the author: ").title()
    book_genre = input("Please enter the genre of the book: ").title()

    # Adding book to the list of books in Library class
    library.add_book(Book(title=book_title,author=author,genre=book_genre))

    library.book_info()

  # Create user
  elif user_input == 2:
    user_name = input("Please enter your name: ")
    if user_name not in users:
      user = User(user_name)
      users.append(user)

  # Search a book by title
  elif user_input == 3:
    if len(library.books)>0:
      book_title = input("Please enter the title of the book: ").title()
      library.search_book(book_title)

    else:
      print("No book available for this criteria.")

  # Borrow a book
  elif user_input == 4:
    if len(users)>0:
      if len(library.books)>0:
        book_title = input("Please enter the title of the book: ").title()
        user.borrow_book(library=library,book_title=book_title)
      else:
        print("No books found.")
    else:
      print("No users found.")

  # Return a book
  elif user_input == 5:
    if len(users)>0:
      user_name = input("Please enter your name: ")
      user_found = False
      for user in users:
        if user.name == user_name:
          user_found = True
          if len(user.borrowed_books)>0:
            book_title = input("Please enter the title of the book: ").title()
            user.return_book(library,book_title)
          else:
            print("No book to return.")
        else:
          print("No users found.")


  # Exit the program
  elif user_input == 6:
    print("Exiting the Library Management System.")
    break

  else:
    print("Invalid option. Please select an option between (1-6).")


```
