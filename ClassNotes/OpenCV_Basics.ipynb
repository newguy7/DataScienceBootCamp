```python
# opencv reading image
import cv2
# for colab we need google.colab.patches
from google.colab.patches import cv2_imshow

image_array = cv2.imread('OIP.jpg')
print(image_array.shape)
print(type(image_array))

# shape will return height, width, channels
# opencv load the images as BGR
```

```python
# total pixel value of the image is height * width; 355 * 474
print(image_array[:,:,2])
```

```python
# display the image
# for other IDE such as VS code, pycharm
# cv2.imshow("image",image_array)
# cv2.waitkey(0)

# for google colab
cv2_imshow(image_array)
# if we put 0 as the value for weight, it will stay and not disappear
# if 5 then will stay for 5 secs and disappear
cv2.waitKey(0)
```

```python
# color change from BGR to RGB
image_rgb = cv2.cvtColor(image_array,cv2.COLOR_BGR2RGB)
#display the image
cv2_imshow(image_rgb)
cv2.waitKey(0)
```

```python
# gray scale image
img_gray = cv2.imread('OIP.jpg',cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
# gray scale image has only 1 channel

#display the image
cv2_imshow(img_gray)
cv2.waitKey(0)
```

```python
# Saving an Image
#(filename, imagefile)
cv2.imwrite("grayscaled.jpg",img_gray)
```

```python
B, G, R = cv2.split(image_array)
```

```python
# Adding two images over one another
img1 = cv2.imread('image1.jpg')
img2 = cv2.imread('image2.jpg')
```

```python
#display image1
cv2_imshow(img1)
cv2.waitKey(0)
```

```python
#display image2
cv2_imshow(img2)
cv2.waitKey(0)
```

```python
# alpha = 0. It is used to increase the image contrast
weightedsum = cv2.addWeighted(img1, 0.5, img2, 0.4, 0)
cv2_imshow(weightedsum)
```

```python
# Image Preprocessing
# Resizing the image
# cv2.resize(src, dsize,interpolation)
# dsize=(new_width, new_height),
# interpolation=cv2.INTER_CUBIC
bigger = cv2.resize(image_array,(700,500))
cv2_imshow(bigger)

# we will resize the image only to the size where it wont be blurry. Blurry image means information loss
```

```python
# cv2.erode()
# used to perform erosion on the image
# it erodes away the boundaries of foreground object (Always try to keep foreground in white)
import numpy as np

kernel = np.ones((3,3),np.uint8)
image_erode = cv2.erode(image_array,kernel)
cv2_imshow(image_erode)
```

```python
# blurred image
# take 3*3 matrix blur it and goes all over the image
Gaussian = cv2.GaussianBlur(image_array,(3,3),0)
cv2_imshow(Gaussian)
cv2.waitKey(0)
```

```python
# one of the best algorithm to remove noise
# image_array, some odd value
# removes the sharpness of the image but donot loose the information
# blur the edges of the object
median_image = cv2.medianBlur(image_array,7)
cv2_imshow(median_image)
cv2.waitKey(0)
```

```python
# Adding border
# value = BGR [255,255,255]
# passes the pixel from top, left, bottom and right (here: 10,10,10,10)
image_border = cv2.copyMakeBorder(image_array, 10, 10, 10, 10, cv2.BORDER_CONSTANT, None, value=[0,100,255])
cv2_imshow(image_border)
```
