```python
'''
Create a Employee Management Case Study
Which Should have
1) Add Employee
2) Remove Employee by id
3) Remove employee by name
4) Update employee by id
5) Update employee by name
6) display all
7) Display By ID
8) Exit
'''
# name, salary, position, id, age
# Catch :- All data must be saved in CSV and when user re run code if any data exists should be loaded automatically
```

```python
import pandas as pd

class Employee:

  def __init__(self,name:str,age:int,id:int,position:str,salary:float):
    self.name = name
    self.age = age
    self.id = id
    self.position = position
    self.salary = salary

  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Id: {self.id}")
    print(f"Position: {self.position}")
    print(f"Salary: {self.salary}")
    print("******************")

class EmployeeManagement:

  def __init__(self):
    self.filename = 'employees.csv'
    self.employees = []
    self.load_csv()


  def save_to_csv(self):
    # create dataframe and
    # add values for each employee in employees
    df = pd.DataFrame([[emp.name, emp.age, emp.id, emp.position, emp.salary] for emp in self.employees],
                          columns=['Name', 'Age', 'ID', 'Position', 'Salary'])
    df.to_csv('employees.csv', index=False)

  def load_csv(self):
    try:
      df = pd.read_csv('employees.csv')
      for index, row in df.iterrows():
        self.employees.append(Employee(row['Name'], row['Age'], row['ID'], row['Position'], row['Salary']))
    except FileNotFoundError:
      pass

  def employee_info(self):
    print("Employee Details:")
    for employee in self.employees:
      employee.display_info()

  def id_exists(self, id):
    for emp in self.employees:
      if emp.id == id:
        return True
    return False

  def add_employee(self,emp):
    return self.employees.append(emp)

  def remove_employee_by_id(self,id):
    employee_to_remove = [emp for emp in self.employees if int(id) == emp.id]
    if employee_to_remove:
      for employee in employee_to_remove:
        employee.display_info()
        self.employees.remove(employee)
        print("*******Employee removed*******")
    else:
      print("No Employee found matching the search criteria.")

  def remove_employee_by_name(self,name):
    employee_to_remove = [emp for emp in self.employees if name.lower() == emp.name.lower()]
    if employee_to_remove:
      for employee in employee_to_remove:
        employee.display_info()
        self.employees.remove(employee)
        print("*******Employee removed*******")
    else:
      print("No Employee found matching the search criteria.")

  def update_employee_by_id(self):
    id = input("Please enter the id to search: ")
    employee_to_update = [emp for emp in self.employees if int(id) == emp.id]
    print("Searching for the employee.....")
    if employee_to_update:
      for emp in employee_to_update:
        print("Current details:")
        emp.display_info()
        emp.position = input("Please enter new position: ")
        emp.salary = float(input("Please enter new salary: "))
        print("-----------------------")
        emp.display_info()
    else:
      print("No Employee found matching the search criteria.")

  def update_employee_by_name(self):
    name = input("Please enter the name to search: ").lower()
    employee_to_update = [emp for emp in self.employees if emp.name.lower() == name]
    if employee_to_update:
      for emp in employee_to_update:
        print("Current details:")
        emp.display_info()
        emp.position = input("Please enter new position: ")
        emp.salary = float(input("Please enter new salary: "))
        print("-----------------------")
        emp.display_info()
    else:
      print("No Employee found matching the search criteria.")

  def display_all(self):
    if len(self.employees != 0):
      for employee in self.employees:
        employee.display_info()
    else:
      print("No employees record found.")

  def display_by_id(self):
    id = input("Please enter the id to search: ")
    employee_to_search = [emp for emp in self.employees if emp.id == int(id)]
    if employee_to_search:
      for emp in employee_to_search:
        emp.display_info()
    else:
      print("No Employee found matching the search criteria.")


emp_manager = EmployeeManagement()

if len(emp_manager.employees)>0:
  emp_manager.employee_info()


while True:
  print("----------------------------------")
  print("Welcome to the Employee Management Case Study")
  print("\nMenu:")
  print("1. Add Employee")
  print("2. Remove Employee by ID")
  print("3. Remove Employee by Name")
  print("4. Update Employee by ID")
  print("5. Update Employee by Name")
  print("6. Display all Employees")
  print("7. Display employee by ID")
  print("8. Exit")
  print("**************************************")

  user_input = int(input("Please set a option from 1 to 8: "))
  if user_input == 1:
    # Adding new employee
    name = input("Please enter the name of the employee: ").title()
    while True:
      try:
        id = int(input("Please enter the id of the employee: "))
        if emp_manager.id_exists(id=id):
          print("Same ID cannot be Used again.")
          continue
        break
      except ValueError:
        print("Invalid ID format. Please enter a valid integer.")

    age = int(input("Please enter the age of the employee: "))

    while True:
      try:
        position = input("Please enter the position of the employee: ")
        if position == "":
          print("Position cannot be let empty.")
          continue
        break
      except ValueError:
        print("Invalid input for position.")

    while True:
      try:
        salary = float(input("Please enter the salary of the employee: "))
        if age <= 0:
          print("Salary must be a positive number.")
          continue
        break
      except ValueError:
        print("Invalid input for salary. Please enter valid number.")

    # Adding employee to the list of employees in EmployeeManagement Class
    emp_manager.add_employee(Employee(name=name,age=age,id=id,position=position,salary=salary))


  # Remove Employee record by ID
  elif user_input == 2:
    if len(emp_manager.employees)>0:
      id = input("Please enter the ID of the employee to remove: ")
      emp_manager.remove_employee_by_id(id=id)
    else:
      print("No Employee data available.")

    emp_manager.employee_info()

  # Remove Employee record by name
  elif user_input == 3:
    if len(emp_manager.employees)>0:
      name = input("Please enter the name of the employee to remove: ")
      emp_manager.remove_employee_by_name(name=name)
    else:
      print("No Employee data available.")

    emp_manager.employee_info()

  # Update Employee details by ID
  elif user_input == 4:
    if len(emp_manager.employees)>0:
      emp_manager.update_employee_by_id()
    else:
      print("No employee data available.")

  # Update Employee details by name
  elif user_input == 5:
    if len(emp_manager.employees)>0:
      emp_manager.update_employee_by_name()
    else:
      print("No employee data available.")

  # Display all Employees
  elif user_input == 6:
    if len(emp_manager.employees)>0:
      emp_manager.employee_info()
    else:
      print("No employee data available.")

  # Display employee by ID
  elif user_input == 7:
    if len(emp_manager.employees)>0:
      emp_manager.display_by_id()
    else:
      print("No Employee data available.")

  # Exit the program
  elif user_input == 8:
    emp_manager.save_to_csv()
    print("Exiting the Employee Management Case Study.")
    break

  else:
    print("Invalid option. Please select an option between (1-8).")

```
