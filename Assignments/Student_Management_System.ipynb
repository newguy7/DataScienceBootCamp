```python
class Student:

  def __init__(self,name:str,rollnumber:int,age:int,marks:float):
    self.name = name
    self.rollnumber = rollnumber
    self.age = age
    self.marks = marks

  def display_info(self):
    print("List of current students:")
    print(f"Name: {self.name}")
    print(f"Rollnumber: {self.rollnumber}")
    print(f"Age: {self.age}")
    print(f"Marks: {self.marks}")
    print("******************")


class StudentManager:

  def __init__(self):
    self.students = []

  def student_info(self):
    for student in self.students:
      student.display_info()

  def add_student(self,student):
    return self.students.append(student)

  def roll_number_exists(self, roll_number):
    for student in self.students:
      if student.rollnumber == roll_number:
        return True
    return False

  def name_exists(self, name):
    for student in self.students:
      if student.name == name:
        return True
    return False

  def remove_student(self,name,roll_number):
    student_to_remove = [student for student in self.students if name.lower() in student.name.lower() and \
                         student.rollnumber == roll_number]
    if student_to_remove:
      print("Search Result.")
      for student in student_to_remove:
        student.display_info()
        self.students.remove(student)
        print("*******Student removed*******")
    else:
      print("No Student found matching the search criteria.")

  def search_student(self,key_word):
    search_results = [student for student in self.students if \
                      student.name.lower() == key_word.lower() or \
                      student.rollnumber == int(key_word)]

    print("Searching for the student.....")
    if search_results:
      print("Search Result.")
      for student in search_results:
        student.display_info()
        print("-----------------------")
    else:
      print("No Student found matching the search criteria.")

  def update_details(self,name,roll_number):
    search_result = [student for student in self.students if \
                      student.name.lower() == name and \
                      student.rollnumber == roll_number]

    print("Searching for the student.....")
    if search_result:
      print("Search Result:")
      for student in search_result:
        student.display_info()
        update_age = int(input("Please enter the updated age for the student: "))
        update_marks = float(input("Please enter the new mark for the student: "))
        student.age= update_age
        student.marks = update_marks
        print("-----------------------")
        student.display_info()
    else:
      print("No Student found matching the search criteria.")


# ----------------------------------------------------------------
# Create instance of StudentManager Class
st_manager = StudentManager()

while True:
  print("----------------------------------")
  print("Welcome to the Student Management System")
  print("\nMenu:")
  print("1. Add a new student record")
  print("2. Remove a new student record")
  print("3. Search for a student by name or roll number")
  print("4. Update student details (age and marks)")
  print("5. Display student list")
  print("6. Exit")
  print("**************************************")

  user_input = int(input("Please set a option from 1 to 6: "))
  if user_input == 1:
    # Adding new students
    name = input("Please enter the name of the student: ").title()
    while True:
      try:
        roll_number = int(input("Please enter the roll number of the student: "))
        if st_manager.roll_number_exists(roll_number=roll_number):
          print("Same Roll Number cannot be Used again.")
          continue
        break
      except ValueError:
        print("Invalid roll number format. Please enter a valid integer.")
    while True:
      try:
        age = int(input("Please enter the age of the student: "))
        if age <= 0:
          print("Age must be a positive integer.")
          continue
        break
      except ValueError:
        print("Invalid input for age. Please enter valid integers.")
    while True:
      try:
        marks = float(input("Please enter the marks of the student: "))
        if marks < 0 or marks > 100:
          print("Marks must be between 0 and 100")
          continue
        break
      except ValueError:
        print("Invalid input for marks. Please enter valid floats.")

    # Adding student to the list of students in StudentManager class
    st_manager.add_student(Student(name=name,rollnumber=roll_number,age=age,marks=marks))

    st_manager.student_info()

  # Remove student record
  elif user_input == 2:
    if len(st_manager.students)>0:
      name = input("Please enter name of the student to remove: ").title()
      roll_number = int(input("Please enter roll number of the student to remove: "))
      st_manager.remove_student(name=name,roll_number=roll_number)
    else:
      print("No student data available.")

    st_manager.student_info()

  # Search for a student by name or roll number
  elif user_input == 3:
    if len(st_manager.students)>0:
      key_word = input("Please enter name or roll number to search the student: ")
      st_manager.search_student(key_word)
    else:
      print("No student data available.")

  # Update student details
  elif user_input == 4:
    if len(st_manager.students)>0:
      name = input("Please enter the name of the student: ").lower()
      roll_number = int(input("Please enter the roll number of the student: "))
      st_manager.update_details(name,roll_number)
    else:
      print("No student data available.")

  # Display Student list
  elif user_input == 5:
    if len(st_manager.students)>0:
      st_manager.student_info()
    else:
      print("No student data available.")

  # Exit the program
  elif user_input == 6:
    print("Exiting the Student Management System.")
    break

  else:
    print("Invalid option. Please select an option between (1-6).")




```
